```python
# Central Difference + Forward Euler

from pylab import*
from scipy.sparse import diags
import numpy as np
# Setup
clf()
dx = 0.01
dt = 0.1
pe = 10

X = r_[0.:3.14:dx]
N = len(X)
B = asmatrix(diags(array([-(pe*dx+1)*ones(N-2), (pe*dx+2)*ones(N-1), -1.*ones(N-2)]), [-1, 0, 1]).toarray())

u = asmatrix(np.cos(X)).T
u[0] = 1.
u[-1] = -1


# System of ODEs ???
f = lambda u: vstack([0., -1*B*u[1::]])
# Explicit Euler
t_final = 4
time = r_[0.:t_final+dt:dt]
pt = array([0., .12, .25, .375])
pn = len(pt)
pc = 0
rt = asmatrix(zeros([1, pn]))
S = asmatrix(zeros([N, pn]))

for t in time:
    # plot storage
    try:
        if (t >= pt[pc]):
            S[:,pc] = u
            rt[0,pc] = t
            pc += 1
            if (pc > pn):
                break
    except:
        pass
        # time advancement
    u = u + dt*f(u)

# Plot unstable figure
figure(1)
linstyl = ["-", "--", ":", "-.", ".-"]
for i in range(pn):
    plot(X, S[:,i], "k%s" % linstyl[i], lw = 1., label = "t = %.2f" % rt[0,i])
xlim([0., 3.14]); #ylim([-4., 4.])
ax = gca()
#ax.set_xticks([0., .25, .5, .75])
#ax.set_yticks([-4., -2., 0., 2., 4.])
grid(axis="both")
xlabel("x", fontsize = 14)
ylabel(r'$\phi$', fontsize=14)
legend(loc = "upper left") # Central Difference + Forward Euler
```
